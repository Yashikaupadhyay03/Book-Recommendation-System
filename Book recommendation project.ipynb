# **POPULARITY BASED RECOMMENDER**

```python
rwithn=rating.merge(books, on = 'ISBN')
```

```python
num_of_rating= rwithn.groupby('Book-Title').count()['Book-Rating'].reset_index()
num_of_rating.rename(columns = {'Book-Rating': 'num_rating'}, inplace= True)
```

```python
avg_of_rating= rwithn.groupby('Book-Title').mean()['Book-Rating'].reset_index()
avg_of_rating.rename(columns = {'Book-Rating': 'avg_rating'}, inplace= True)
avg_of_rating
```

```python
popular = num_of_rating.merge(avg_of_rating, on = 'Book-Title')
popular
```

```python
popular[popular['num_rating']>=250]
```

```python
popular = popular[popular['num_rating']>=250].sort_values('avg_rating', ascending = False).head(50)
```

# **COLLABORATIVE FILTERING BASED RECOMMENDER**

```python
rwithn
```

now we will check on how many books a user have left a rating

```python
x = rwithn.groupby('User-ID').count()['Book-Rating'] > 200
users_with_200_rating = x[x].index
```

```python
rwithn['User-ID'].isin(users_with_200_rating)
```

```python
filtered_rating= rwithn[rwithn['User-ID'].isin(users_with_200_rating)]
```

```python
#now on the basis of book title

y = filtered_rating.groupby('Book-Title').count()['Book-Rating']>=50
famous_books= y[y].index
```

```python
filtered_rating['Book-Title'].isin(famous_books)
```

```python
final_rating = filtered_rating[filtered_rating['Book-Title'].isin(famous_books)]
```

```python
final_rating.drop_duplicates()
```

```python
pt = final_rating.pivot_table(index = 'Book-Title', columns = 'User-ID', values = 'Book-Rating')
```

```python
pt
```

```python
pt.fillna(0, inplace = True)
pt
```

Now we will find the euclidean distances between all the books and then the books that have less distance between them are same.

```python
from sklearn.metrics.pairwise import cosine_similarity as cs
```

```python
cs(pt)
```

```python
simarlity_score = cs(pt)
```

```python
simarlity_score[0]
#distance of book 1 with every other book
```

```python
#def recommend(book_name):
  #index fetch
  #return suggestions
```

```python
#HOW TO FETCH INDEX
np.where(pt.index== 'A Bend in the Road')[0][0]
```

```python
#HOW TO FIND SIMILARITY SCORE OF BOOK WITH ALL OTHER BOOKS
sorted(list(enumerate(simarlity_score[0])), key = lambda x:x[1], reverse = True)[1:6]
#SABSE JYADA SIMILAR SBSE UPPAR HAI. PHELA VALA VOH BOOK KHUD HAI TOH 1 SE START.
```

```python
def recommend(book_name):
  #index fetch
  index = np.where(pt.index== book_name)[0][0]
  similar_books = sorted(list(enumerate(simarlity_score[index])), key = lambda x:x[1], reverse = True)[1:6]

  for i in similar_books:
    print(pt.index[i[0]])
```

```python
recommend('Whispers')
```
